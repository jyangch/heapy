```python
import numpy as np
from astropy.io import ascii
from heapy.autobs.polybase import PolyBase
```

```python
LC = ascii.read('./lc.txt')
time = LC['time']
rate = LC['rate']

binsize = time[1] - time[0]
bins = time - binsize / 2
bins = np.append(bins, (time + binsize / 2)[-1])
cts = np.array(rate * binsize).astype(int)
```

```python
bs = PolyBase.frombin(cts, bins)

bs.basefit()
bs.bblock()
bs.calsnr()
bs.sorting()
bs.polyfit()

bs.savedata('./autobs')
bs.savefig('./autobs', show=True)
```

```python
bs.loop()
bs.savedata('./autobs')
bs.savefig('./autobs', show=True)
```
