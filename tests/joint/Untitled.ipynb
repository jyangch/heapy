```python
group = {
    'src': '../wxt/spectrum/p0d00-p160d00.src', 
    'bkg': '../wxt/spectrum/p0d00-p160d00.bkg', 
    'rmf': '../wxt/ep.rmf', 
    'arf': '../wxt/ep.arf'}

spec = Spectrum()
spec.set('wxt', group, nt=[0.5, 4], stat='cstat', gr={'min_evt': 5})

tbabs_ = tbabs()
tbabs_.expr = 'tbabs'
tbabs_.redshift = 0

pl_ = pl()

mo = Model()
mo.set(expr='tbabs * pl', obje=[tbabs_, pl_])

fit = Fit()
fit.set(spec, mo)
fit.run('multinest', './wxt_msf_nh', nlive=400)

ana = Analyse(fit)
ana.post()
ana.corner()

plot = Plot(ana)
plot.rebin()
plot.cspec()
plot.pspec()
```
