```python
import numpy as np
from astropy.io import ascii
from heapy.temporal.txx import pgTxx
```

```python
LC = ascii.read('./lc.txt')
time = LC['time']
rate = LC['rate']

binsize = time[1] - time[0]
bins = time - binsize / 2
bins = np.append(bins, (time + binsize / 2)[-1])
cts = np.array(rate * binsize).astype(int)
```

```python
txx = pgTxx.frombin(cts, bins)
txx.accumcts(mp=False)
txx.savedata('./txx')
txx.savefig('./txx', show=True)
```
